# Indian Restaurant Market Analysis in Germany

## Notebook Goal

This notebook analyzes Yelp data to evaluate the feasibility of launching an Indian food truck in Germany, with a primary focus on Berlin.

## Main Objectives

- Load and inspect the Yelp dataset  
- Identify relevant business and review information  
- Filter restaurant-related data  
- Prepare the dataset for analysis  
- Analyze patterns related to competition, ratings, and customer feedback  

## 1. Project Setup

### Project Context

This project evaluates the feasibility of launching an Indian food truck in Germany using a data-driven market research approach.

The analysis is structured into two complementary components:

- **Market & Location Analysis**  
  Focused on demand, demographics, and location potential.

- **Restaurant & Competition Analysis (this notebook)**  
  Focused on existing restaurant data using Yelp, including competition levels, customer ratings, pricing signals, and feedback.

This notebook specifically contributes to understanding the competitive landscape and customer perception of Indian food in the market.

---

### Role of This Notebook

This notebook focuses on analyzing the competitive landscape of the restaurant market using Yelp data.

The objective is to understand how Indian food is currently positioned in the market by evaluating:
- competition density
- customer satisfaction (ratings and reviews)
- pricing signals
- recurring customer feedback

This analysis aims to identify patterns and potential gaps that could represent opportunities for a food truck concept.

---

### Key Analytical Questions

- How strong is the competition for Indian food?
- What are the current customer satisfaction levels?
- What do customers like or dislike about existing restaurants?
- Are there unmet needs that a food truck could address?

### Data Source

The primary dataset used in this notebook is the Yelp dataset, which includes:

- business information (name, location, categories)
- ratings (stars)
- number of reviews
- (optional) customer review text

This dataset enables both quantitative analysis (ratings, counts) and qualitative insights (customer feedback).

---

### Analytical Objective

The goal of this notebook is to extract insights about:
- competition density
- customer perception
- pricing and positioning signals

These insights will contribute to the overall business decision of whether launching an Indian food truck is viable, and how it should be positioned in the market.

```python
# Define the path to the business dataset
business_path = dataset_folder / "yelp_academic_dataset_business.json"

# Check if the file exists
print("Business file exists:", business_path.exists())
print("Business file path:", business_path)
```

```python
import json
import pandas as pd

# Load the Yelp business dataset
business_data = []

with open(business_path, "r", encoding="utf-8") as file:
    for line in file:
        business_data.append(json.loads(line))

# Convert the raw data into a DataFrame
business_df = pd.DataFrame(business_data)

# Preview the dataset
business_df.head()
```
